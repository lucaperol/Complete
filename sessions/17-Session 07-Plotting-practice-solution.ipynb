### For each of the following tasks create a graph

### **Task 1**. Is there any significant relationship between life expectancy and GDP in the UK ?
Tip: we need to create a scatter plot

```python
df_uk=df[df['country']=='United Kingdom']
ax = sns.regplot(x='lifeExp', y='gdpPercap', data=df_uk)
ax.set_title('Life Expectency & GDP per capita');
ax.set_xlabel('GDP per capita');
ax.set_ylabel('Life Expectency');
```

### **Task 2**. Plot the number of observations for each continent
tip we need to create a bar chart (countplot)

```python
ax = sns.countplot(x='continent', data=df)
ax.set_title('Observations in different countries');
ax.set_xlabel('Continent');
ax.set_ylabel('Observations');
```

### **Task 3**. Plot the evolution of life expectancy across different contintent

```python
sns.lineplot(data = df,x="year", y="lifeExp", hue='continent')
```

### **Task 4**. Plot the distribution of lifeExp for different years and different continents using facetgrid

```python
facet = sns.FacetGrid(df, col='year', hue='continent')
facet = facet.map(plt.hist, 'lifeExp')
facet = facet.add_legend()
```

### **Task 5**. Use pairplot to compare variables across continents

```python
fig = sns.pairplot(df,hue='continent')
```

### **Task 6**. Boxplot for continent population for year 1980-1990 inclusive

```python
plt.figure(figsize=(16, 8))
ax = sns.boxplot(x='continent', y='pop', data=df[(df['year']>=1980) & (df['year']<=1990)])
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```
